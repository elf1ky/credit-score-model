```python
#if data is object
#data['column_name'].fillna('Unknown', inplace=True)
#if numeric
data['person_emp_length'] =data['person_emp_length'].fillna(data['person_emp_length'].mean())
data['loan_int_rate'] =data['loan_int_rate'].fillna(data['loan_int_rate'].mean())
```

```python
import numpy as np

# Detect outliers for each numerical column using the IQR method
numeric_cols = data.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]

    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
#loan_grade with label encoder the rest with one hot encodeing
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
data['loan_grade']= le.fit_transform(data['loan_grade'])
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False, drop='first')  # remove drop='first' if you want all categories

# Select the categorical columns
categorical_cols = ['person_home_ownership', 'loan_intent', 'cb_person_default_on_file']

# Fit and transform the data
encoded_data = encoder.fit_transform(data[categorical_cols])

# Create a DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate with the original DataFrame (excluding the original categorical columns)
df_encoded = pd.concat([data.drop(columns=categorical_cols), encoded_df], axis=1)
```

```python
df_encoded.info()
```

```python
from sklearn.model_selection import train_test_split
x=df_encoded.drop('loan_status', axis=1)
y=df_encoded['loan_status']
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
print("Train columns:", x_train.columns)
print("Test columns:", x_test.columns)
#making sure train and test columns are the same after splitting and nothing is missing
```

```python
print(set(x_train.columns) - set(x_test.columns))  # what's in train but not in test
print(set(x_test.columns) - set(x_train.columns))  # what's in test but not in train
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
#after using grid search these were the best hyper parameters
forest = RandomForestClassifier(n_estimators=100 ,criterion= 'gini', max_depth= None,random_state= 42)

forest.fit(x_train, y_train)
# Predict on test set
y_pred= forest.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
#accuracy
accuracy= accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy*100,"%")
```

```python
#Feature scaling isn’t necessary for RandomForest
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [10, 50, 100],# Number of trees in the forest
    'max_depth': [None, 10, 20], # Maximum depth of each tree(None means unlimited depth).
    'criterion': ['gini', 'entropy'] # Split criteria for decision trees
}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='accuracy')
#cv=5 -> 5-fold cross-validation is used (splitting data into 5 parts, training on 4, and testing on 1).
#scoring='accuracy -> The performance metric used to evaluate each set of hyperparameters.
grid_search.fit(x_train, y_train)
#This trains multiple RandomForestClassifier models using different combinations of hyperparameters from param_grid.

#Each model is evaluated using cross-validation, and the best combination is selected.
print(grid_search.best_params_)
```
